```python
import threading
import time

class PingPong(threading.Thread):
    def __init__(self, word):
        super().__init__()
        self.word = word   

    def run(self):
        for i in range(100):
            print(self.word, end='', flush=True)
            time.sleep(0.001)

def detect_concurrency():
    def trace_thread(frame, event, arg):
        if event == 'call':
            print(f"[TRACE] Llamando: {frame.f_code.co_name} en {frame.f_lineno}")
        return trace_thread
    threading.settrace(trace_thread)

if __name__ == '__main__':
    detect_concurrency()
    ping = PingPong('ping')
    pong = PingPong('pong')
    ping.start()
    pong.start()
    ping.join()
    pong.join()
    print('\nFin')
    threading.settrace(None)
```

```python
import threading
import random
import time

class Carrera4x100:
    testigo = threading.Semaphore(1)
    tiempo_inicio = None

    def __init__(self, atleta):
        self.atleta = atleta

    def correr(self):
        # Esperar el testigo para correr
        Carrera4x100.testigo.acquire()
        if Carrera4x100.tiempo_inicio is None:
            Carrera4x100.tiempo_inicio = time.time()

        print(f"Atleta {self.atleta} comienza a correr")
        tiempo_carrera = random.uniform(9, 11)
        time.sleep(tiempo_carrera)
        print(f"Atleta {self.atleta} ha terminado en {tiempo_carrera:.2f} segundos")

        Carrera4x100.testigo.release()

if __name__ == "__main__":
    atletas = [Carrera4x100(i) for i in range(1, 5)]

    hilos = [threading.Thread(target=atleta.correr) for atleta in atletas]

    for hilo in hilos:
        hilo.start()
        hilo.join()

    tiempo_total = time.time() - Carrera4x100.tiempo_inicio
    print(f"Tiempo total de la carrera: {tiempo_total:.2f} segundos")
```
