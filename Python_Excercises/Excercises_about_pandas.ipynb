### Assignment 4: Data Aggregation

1. Create a Pandas DataFrame with 2 columns: 'Category' and 'Value'. Fill the 'Category' column with random categories ('A', 'B', 'C') and the 'Value' column with random integers. Group the DataFrame by 'Category' and compute the sum and mean of 'Value' for each category.
2. Create a Pandas DataFrame with 3 columns: 'Product', 'Category', and 'Sales'. Fill the DataFrame with random data. Group the DataFrame by 'Category' and compute the total sales for each category.

```python
categories = ['A', 'B', 'C']
values=np.random.randint(1,199, size=10)
df = pd.DataFrame({'categories': np.random.choice(categories, 10), 'values': values})
df2 = df.groupby('categories')['values'].agg(['mean', 'sum'])
df2
```

```python
products = ['Laptop', 'Phone', 'Tablet', 'Headphones', 'Smartwatch', 'Monitor']
categories = ['Electronics', 'Accessories']
sales = np.random.randint(100, 1000, size=len(products)) 

df = pd.DataFrame({
    'Product': products,
    'Category': np.random.choice(categories, size=len(products)),  
    'Sales': sales
})
df2 = df.groupby('Category')['Sales'].agg(['sum'])
df2
```

### Assignment 5: Merging DataFrames

1. Create two Pandas DataFrames with a common column. Merge the DataFrames using the common column.
2. Create two Pandas DataFrames with different columns. Concatenate the DataFrames along the rows and along the columns.

```python

df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 45]
})
df2 = pd.DataFrame({
    'ID': [3, 4, 5, 6, 7],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'Salary': [60000, 70000, 80000, 90000, 100000]
})
df3=pd.merge(df1,df2, on='ID',how='left')
df3
```

```python

df1 = pd.DataFrame({
    'StudentID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Grade': ['A', 'B', 'A', 'C']
})
df2 = pd.DataFrame({
    'CourseID': [101, 102, 103, 104],
    'CourseName': ['Math', 'Science', 'History', 'English'],
    'Credits': [3, 4, 3, 2]
})

df3 =pd.merge(df1,df2, left_index=True, right_index=True)
df3
```


### Assignment 6: Time Series Analysis

1. Create a Pandas DataFrame with a datetime index and one column filled with random integers. Resample the DataFrame to compute the monthly mean of the values.
2. Create a Pandas DataFrame with a datetime index ranging from '2021-01-01' to '2021-12-31' and one column filled with random integers. Compute the rolling mean with a window of 7 days.

```python
date_range = pd.date_range(start='2024-01-01', periods=100, freq='D')
df = pd.DataFrame({
    'Date': date_range,
    'Sales': np.random.randint(100, 1000, size=len(date_range))
})
df.resample('ME', on='Date').sum()
```

```python
date_range = pd.date_range(start='2021-01-01', end='2021-12-31', freq='D')
df = pd.DataFrame({
    'Date': date_range,
    'Sales': np.random.randint(100,1000, size=len(date_range))
})
df.resample('W', on='Date').mean()
```


### Assignment 8: Pivot Tables

1. Create a Pandas DataFrame with columns 'Date', 'Category', and 'Value'. Create a pivot table to compute the sum of 'Value' for each 'Category' by 'Date'.
2. Create a Pandas DataFrame with columns 'Year', 'Quarter', and 'Revenue'. Create a pivot table to compute the mean 'Revenue' for each 'Quarter' by 'Year'.

```python
data_range= pd.date_range(start='2021-01-01', end='2021-12-31', freq='D')
categories = ['Electronics', 'Clothing']
values = np.random.randint(100, 1000, size=len(data_range))
df=pd.DataFrame({'Date': data_range, 'Category': np.random.choice(categories, len(data_range)), 'Sales': values})
df.set_index('Date', inplace=True)
pd.pivot_table(df, index='Date', columns='Category', values='Sales', aggfunc='sum')
```

```python

years = np.repeat([2022, 2023, 2024], 4)  
quarters = ['Q1', 'Q2', 'Q3', 'Q4'] * 3  
revenues = np.random.randint(10000, 50000, size=12)  

df = pd.DataFrame({
    'Year': years,
    'Quarter': quarters,
    'Revenue': revenues
})

pd.pivot_table(df, values='Revenue', index='Year', columns='Quarter', aggfunc='mean')
```
